```python
#*********************************************
# Levi Sutton
# CS 4412 (Tuesday/Thursday 6:30)
# Homework 1 (Part 2)
#
#*********************************************

# Import packages
# Note: Install apriori (pip install apyori)
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori

data = pd.read_excel("Online Retail.xlsx") # reads data from the excel sheet, data = pd.read_excel("https://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx")
df = pd.DataFrame(data) # converts the data from excel sheet into a data frame for easier usability and access Data Frame attributes
df1 = df.iloc[:,[0,1,2,7]] # selects the range of data from all rows combined with the 1st, second, third and 8th columns
df2 = df1[df1.Country.str.contains("United Kingdom")]# filters out the data that contains United Kingdom
print(df2)# Print out filtered data

association_rules = apriori(df2, min_support=0.1, min_confidence=0.2, min_lift=3, min_length=2)# use the apriori method to gather association rules for our data
association_results = list(association_rules) # puts association rules in a list

print()
print("Association Rules")
print()

# Loop to generate the results/Candiate Generations for the association rules
for associations in association_results:
    
    pairs = associations[0] 
    items = [x_item for x_item in pairs]
    
    print("Rule: " + items[0] + " -> " + items[1])
    print("Support: " + str(associations[1]))
    print("Confidence: " + str(associations[2][0][2]))
    print("_______________________________________")
```
